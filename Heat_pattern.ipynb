```
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
n=int(input())
a=np.ones((n,n))
for k in range(1,(n//2)+1):
	if k%2!=0:
		a[k:n-k,k:n-k]=np.zeros((n-2*k,n-2*k))
	else:
		a[k:n-k,k:n-k]=np.ones((n-2*k,n-2*k))
		
%matplotlib inline
plt.figure(figsize=(16,8))
sns.heatmap(a)
```
